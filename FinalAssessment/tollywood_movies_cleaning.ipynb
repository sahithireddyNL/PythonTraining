## Loading the files and Importing pandas

```python
import pandas as pd
df = pd.read_excel("tollywood_movies.xlsx")
df
```

```python
df.info()
```

```python
# Removing column as it was of no use
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
# Filling missing Language with 'Telugu'
df['Language'] = df['Language'].fillna('Telugu')
df
```

```python
# Droping rows where ReleaseYear is NaN
df = df.dropna(subset=['ReleaseYear'])
df
```

```python
# 1. Removing duplicate Titles, keeping the ones with fewer NaNs
def get_row_with_fewest_nans(group):
    return group.isnull().sum(axis=1).idxmin()

df = df.loc[df.groupby('Title').apply(get_row_with_fewest_nans)]
df
```

```python
# Calculating the overall average rating
overall_mean_rating = df['Rating'].mean()

# Filling missing Rating with the overall average
df['Rating'] = df['Rating'].fillna(overall_mean_rating).round(1)
df
```

```python
# Converting ReleaseYear from float to int for better understanding
df['ReleaseYear'] = df['ReleaseYear'].astype(int)
df
```

```python
# Calculating the mean of the 'Duration (minutes)' column
mean_duration = df['Duration (minutes)'].mean()

# Filling missing values in 'Duration (minutes)' with the calculated mean
df['Duration (minutes)'] = df['Duration (minutes)'].fillna(mean_duration).astype(int)
df
```

```python
# Dropping rows with NaN values in Column 'ProductionCompany'
df = df.dropna(subset=['ProductionCompany'])
df
```

```python
# Saving the cleaned data to a csv file
df.to_csv('tollywood_movies_cleaned.csv', index=False)
```
