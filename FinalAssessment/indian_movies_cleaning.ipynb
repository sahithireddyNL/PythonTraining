## Loading the files and importing pandas

```python
import pandas as pd
df = pd.read_csv("indian_movies.csv")
df
```

```python
df.info()
```

```python
# Removing column as it was of no use
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
# Droping rows where ReleaseYear or Rating is NaN
df = df.dropna(subset=['ReleaseYear', 'Rating'])
df
```

```python
# Converting ReleaseYear from float to int for better understanding
df['ReleaseYear'] = df['ReleaseYear'].astype(int)
df
```

```python
# Filling NaNs in Budget and BoxOffice with aggregate function mean
df.loc[:, 'Budget (Crores)'] = df['Budget (Crores)'].fillna(df['Budget (Crores)'].mean())
df.loc[:, 'BoxOffice (Crores)'] = df['BoxOffice (Crores)'].fillna(df['BoxOffice (Crores)'].mean())
df
```

```python
# Calculate the mean of the 'Duration (minutes)' column
mean_duration = df['Duration (minutes)'].mean()

# Fill missing values in 'Duration (minutes)' with the calculated mean
df['Duration (minutes)'] = df['Duration (minutes)'].fillna(mean_duration).astype(int)
df
```

```python
# Saving the cleaned data to a csv file
df.to_csv('indian_movies_cleaned.csv', index=False)
```
