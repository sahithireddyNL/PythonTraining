## Plot average BoxOffice Collection by Genre for combined dataset
What is your inference from this?

How does average change when we pick only Bollywood movies?

What is your inference now?


```python
import pandas as pd
import matplotlib.pyplot as plt

indian_movies = pd.read_csv('indian_movies_cleaned.csv')
bollywood = pd.read_csv('bollywood_movies_cleaned.csv')
tollywood = pd.read_csv('tollywood_movies_cleaned.csv')

combined = pd.concat([indian_movies, bollywood, tollywood], ignore_index=True)
combined.columns = combined.columns.str.strip()
combined['BoxOffice (Crores)'] = pd.to_numeric(combined['BoxOffice (Crores)'], errors='coerce')

combined.drop_duplicates(subset='Title', inplace=True)

def map_genre(genre):
    genre = genre.lower()  
    if 'action' in genre:
        return 'Action'
    elif 'drama' in genre:
        return 'Drama'
    elif 'comedy' in genre:
        return 'Comedy'
    elif 'romance' in genre:
        return 'Romance'
    elif 'thriller' in genre:
        return 'Thriller'
    elif 'adventure' in genre:
        return 'Adventure'
    else:
        return 'Other'

combined['Main Genre'] = combined['Genre'].apply(map_genre)

avg_by_genre_all = combined.groupby('Main Genre')['BoxOffice (Crores)'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(avg_by_genre_all.index, avg_by_genre_all.values, color='skyblue')
plt.title('Average BoxOffice Collection by Genre (All Industries)')
plt.xlabel('Genre')
plt.ylabel('Avg BoxOffice (Crores)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Q3. Average boxofficecollections.png')
plt.show()
```

### Inference For All Movies Average: Action has higher average compared to all other genres and a combination of multiple genres have next best average


```python
import pandas as pd
import matplotlib.pyplot as plt

bollywood = pd.read_csv('bollywood_movies_cleaned.csv')
bollywood.columns = bollywood.columns.str.strip()
bollywood['BoxOffice (Crores)'] = pd.to_numeric(bollywood['BoxOffice (Crores)'], errors='coerce')

bollywood.drop_duplicates(subset='Title', inplace=True)

def map_genre(genre):
    genre = genre.lower()  
    if 'action' in genre:
        return 'Action'
    elif 'drama' in genre:
        return 'Drama'
    elif 'comedy' in genre:
        return 'Comedy'
    elif 'romance' in genre:
        return 'Romance'
    elif 'thriller' in genre:
        return 'Thriller'
    elif 'adventure' in genre:
        return 'Adventure'
    else:
        return 'Other'

bollywood['Main Genre'] = bollywood['Genre'].apply(map_genre)

avg_by_genre_bollywood = bollywood.groupby('Main Genre')['BoxOffice (Crores)'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(avg_by_genre_bollywood.index, avg_by_genre_bollywood.values, color='lightgreen')
plt.title('Average BoxOffice Collection by Genre (Bollywood Only)')
plt.xlabel('Genre')
plt.ylabel('Avg BoxOffice (Crores)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Q3. Average boxofficecollections Bollywood.png')
plt.show()
```
