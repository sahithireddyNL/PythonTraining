## Loading Files and Importing Pandas

```python
# Converting sqlite file to csv for handling date using pandas
import sqlite3
import pandas as pd

conn = sqlite3.connect('bollywood_movies.sqlite')

df = pd.read_sql_query("SELECT * FROM movies", conn)
df.to_csv("bollywood_movies.csv", index=False)
conn.close()
```

```python
# Removing column as it was of no use
df = df.drop(columns=['index'])
df
```

```python
# Find rows with the title 'Lagaan'
lagaan_rows = df[df['Title'] == 'Lagaan']

# Display the rows
lagaan_rows
```

```python
# Function to fill missing values based on the other rows with the same title
def fill_missing_values(group, column):
    # Fill NaNs using forward fill and backward fill
    group[column] = group[column].fillna(method='ffill').fillna(method='bfill')
    return group

# Function to handle filling missing values for multiple columns
def fill_all_missing_values(df, columns):
    for column in columns:
        df = df.groupby('Title', group_keys=False).apply(fill_missing_values, column=column)
    return df

# Function to remove duplicates based on Title, keeping only the row with the filled values
def remove_duplicate_titles(df):
    # Remove duplicates, keeping the first row with filled values
    df = df.drop_duplicates(subset=['Title'], keep='first')
    return df

# Columns that need to be filled
columns_to_fill = ['Genre', 'Budget (Crores)', 'BoxOffice (Crores)']

# Fill missing values for specified columns
df = fill_all_missing_values(df, columns_to_fill)

# Remove duplicate titles, keeping the row with fewer NaNs
df = remove_duplicate_titles(df)
df
```

```python
# Step 1: Identify missing values for Budget and BoxOffice
missing_budget_boxoffice = df[df['Budget (Crores)'].isnull() | df['BoxOffice (Crores)'].isnull()]

# Step 2: Group by Genre and calculate mean Budget and BoxOffice for each genre
mean_values_by_genre = df.groupby('Genre').agg({
    'Budget (Crores)': 'mean',
    'BoxOffice (Crores)': 'mean'
}).reset_index()

# Step 3: Fill missing Budget and BoxOffice values based on Genre
def fill_missing_values(row, mean_values_by_genre):
    genre_mean = mean_values_by_genre[mean_values_by_genre['Genre'] == row['Genre']]
    if pd.isnull(row['Budget (Crores)']):
        row['Budget (Crores)'] = genre_mean['Budget (Crores)'].values[0]
    if pd.isnull(row['BoxOffice (Crores)']):
        row['BoxOffice (Crores)'] = genre_mean['BoxOffice (Crores)'].values[0]
    return row

# Apply the function to rows with missing values
df = df.apply(lambda row: fill_missing_values(row, mean_values_by_genre), axis=1)

df
```

```python
df['Budget (Crores)'] = df['Budget (Crores)'].fillna(0).astype(int)
df['BoxOffice (Crores)'] = df['BoxOffice (Crores)'].fillna(0).astype(int)
df
```

```python
# Saving the cleaned data to a csv file
df.to_csv('bollywood_movies_cleaned.csv', index=False)
```
